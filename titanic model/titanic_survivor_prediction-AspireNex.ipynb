```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic_data = pd.read_csv('archive/titanic.csv')
```

```python
titanic_data.describe()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].mean())
titanic_data['Fare'] = titanic_data['Fare'].fillna(titanic_data['Fare'].mean())
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data['Embarked'] = titanic_data['Embarked'].map( {'Q': 0,'S':1,'C':2}).astype(int)
titanic_data['Sex'] = titanic_data['Sex'].map( {'female': 1,'male':0}).astype(int)
```

```python
data = titanic_data.drop(['PassengerId','Name','Cabin','Ticket'], axis =1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

train = titanic_data.drop(['Survived'], axis=1)
test = titanic_data.iloc[:,1]
x_train, x_test, y_train, y_test = train_test_split(train, test, test_size = 0.2, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

clf = LogisticRegression(solver='liblinear', max_iter=100)
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
score = accuracy_score(y_pred,y_test)
score = round(score, 4)
print(f'Logistic regression accuracy: {score*100}%')
```
